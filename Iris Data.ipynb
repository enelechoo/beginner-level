## Visualisations

### Boxplot of species and variables

```python
sns.boxplot(data=iris).set_title("Comparison of variables distribution")
```

Objective: Distribution of data based on species

Interpretation:
1. Sepal_width has the smallest variation amongst the iris species
2. Petal_length has the largest variation amongst the iris species

#### Compare distirbutions of variables (sepal_length,sepal_width,petal_length,petal_width)

```python
sns.boxplot(x="species",y="sepal_length",data=iris).set_title("Compare the distribution of sepal length")
```

```python
sns.boxplot(x="species",y="sepal_width",data=iris).set_title("Compare the distribution of sepal width")
```

```python
sns.boxplot(x="species",y="petal_length",data=iris).set_title("Compare the distribution of petal length")
```

```python
sns.boxplot(x="species",y="petal_width",data=iris).set_title("Compare the distribution of petal width")
```

Interpretation: Distinct differences between petal_length, petal_width, and sepal_length of iris species

### Violin plots of petal width and petal length

```python
sns.violinplot(x="species",y="petal_width",data=iris)
```

```python
sns.violinplot(x="species",y="petal_length",data=iris)
```

### Pairplot of iris dataset

```python
sns.pairplot(iris,hue="species")
```

Objectives: Explore the interactions between the 4 variables

Interpretation: 
1. Setosa has a significantly smaller petal width and petal length across the species
2. Virginica is the largest species
